## Text preprocessing

The textual data has to be preprocessed before it can be used for machine learning.

```python
import nltk
from nltk.corpus import stopwords 
import re
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

Function for removing the punctuation, stemming and lemmatisation of text.

```python
def preprocess_text(text, flg_stemm=False, flg_lemm=True):

    lst_stopwords = stopwords.words("english")
    
    ## clean (convert to lowercase and remove punctuations and characters and then strip)
    text = re.sub(r'[^\w\s]', '', str(text).lower().strip())
            
    ## Tokenize (convert from string to list)
    lst_text = text.split()
    ## remove Stopwords
    if lst_stopwords is not None:
        lst_text = [word for word in lst_text if word not in 
                    lst_stopwords]
                
    ## Stemming (remove -ing, -ly, ...)
    if flg_stemm == True:
        ps = nltk.stem.porter.PorterStemmer()
        lst_text = [ps.stem(word) for word in lst_text]
                
    ## Lemmatisation (convert the word into root word)
    if flg_lemm == True:
        lem = nltk.stem.wordnet.WordNetLemmatizer()    
        lst_text = [lem.lemmatize(word) for word in lst_text]
            
    ## back to string from list
    text = " ".join(lst_text)
    return text
```

```python
train_test["clean_abstract"] = train_test["ABSTRACT"].apply(lambda x: preprocess_text(x, flg_stemm=False, flg_lemm=True, ))
```

# Data processing for deep learning

Using the text without stopwords for the deep learning since we are interested only in the general topic, not the sentiment etc.

```python
# split the train_test
x_train = train_test[:ntrain]['clean_abstract']
y_train = train[target_cols]
X_test = train_test[ntrain:]['clean_abstract']
```

```python
X_test.shape, x_train.shape, y_train.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

Tokenizing the texts and padding them to make them all the same length.

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

tok = Tokenizer(num_words = 1000000)
tok.fit_on_texts(train_test['clean_abstract'].str.lower().tolist())

vocab_size = len(tok.word_index) + 1
```

```python
X_train = tok.texts_to_sequences(X_train)
X_val = tok.texts_to_sequences(X_val)
X_test = tok.texts_to_sequences(X_test)
```

```python
maxlen = 200
X_train = pad_sequences(X_train, maxlen=maxlen)
X_val = pad_sequences(X_val, maxlen=maxlen)
X_test = pad_sequences(X_test, maxlen=maxlen)
```

## Tensorflow model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense, Dropout, SpatialDropout1D, LSTM, Bidirectional
```

Simple deep neural network with the embedding layer, one hidden dense layer and the output layer that gives probabilities of any given category applying to the given text.

```python
embedding_dim = 50
vocab_size = len(tok.word_index) + 1

model = Sequential()
model.add(Embedding(input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=maxlen))

model.add(Flatten())
model.add(Dense(200, activation='relu', name = 'Fully_Connected'))
model.add(Dense(25, activation='sigmoid', name = 'Output'))
model.compile(optimizer=tf.keras.optimizers.Adam(lr = 1e-3),
              loss='binary_crossentropy',
              metrics=['accuracy'],
              )

model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), verbose=True, epochs=20, batch_size=256,
          callbacks = [tf.keras.callbacks.ReduceLROnPlateau()])
```

Choosing the thresholds that will give the best possible f1 score.

```python
def get_best_thresholds(true, preds):
    thresholds = [i/100 for i in range(100)]
    best_thresholds = []
    for idx in range(25):
        f1_scores = [f1_score(true[:, idx], (preds[:, idx] > thresh) * 1) for thresh in thresholds]
        best_thresh = thresholds[np.argmax(f1_scores)]
        best_thresholds.append(best_thresh)
    return best_thresholds
```

```python
val_preds = model.predict(X_val)
best_thresholds = get_best_thresholds(y_val.values, val_preds)
for i, thresh in enumerate(best_thresholds):
    val_preds[:, i] = (val_preds[:, i] > thresh) * 1
f1_score(y_val, val_preds, average='micro')
```

```python
preds_test = model.predict(X_test)

for i, thresh in enumerate(best_thresholds):
    preds_test[:, i] = (preds_test[:, i] > thresh) * 1
```

```python
submit_file(preds_test, filename='hacklive_deep_learning.csv')
```

# LSTM model

```python
# split the train_test
x_train = train_test[:ntrain]['clean_abstract']
y_train = train[target_cols]
X_test = train_test[ntrain:]['clean_abstract']
```

```python
X_test.shape, x_train.shape, y_train.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

Tokenizing the texts and padding them to make them all the same length.

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

tok = Tokenizer(num_words = 1000000)
tok.fit_on_texts(train_test['clean_abstract'].str.lower().tolist())

word_index = tok.word_index

vocab_size = len(tok.word_index) + 1
```

```python
X_train = tok.texts_to_sequences(X_train)
X_val = tok.texts_to_sequences(X_val)
X_test = tok.texts_to_sequences(X_test)
```

```python
maxlen = 450
X_train = pad_sequences(X_train, maxlen=maxlen, padding='post',truncating='post')
X_val = pad_sequences(X_val, maxlen=maxlen, padding='post',truncating='post')
X_test = pad_sequences(X_test, maxlen=maxlen, padding='post',truncating='post')
```

A more advanced neural network, that utilizes LSTM layers (long short term memory) to better understand the text. The use of LSTMs helps with the problem of vanishing and exploding gradients.

```python
embedding_dim = 128
vocab_size = len(tok.word_index) + 1

model = Sequential([
    tf.keras.layers.Embedding(vocab_size,embedding_dim),
    
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(embedding_dim)),
    #tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(embedding_dim/2)),
    tf.keras.layers.Dropout(.2),
    
    tf.keras.layers.Dense(embedding_dim, activation='relu'),
    
    tf.keras.layers.Dense(25, activation='sigmoid')   
])
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr = 1e-3),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), verbose=True, epochs=25, batch_size=256,
          callbacks = [tf.keras.callbacks.ReduceLROnPlateau()])
```

```python
def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.plot(history.history['val_' + string])
    plt.xlabel('Epochs')
    plt.ylabel(string)
    plt.legend([string, 'val_' + string])
    plt.show()
```

```python
plot_graphs(history, 'accuracy')
```

```python
plot_graphs(history, 'loss')
```

Beyond 10 epochs the model started overfitting since the accuracy on the training set kept improving but not on the validation set.

Choosing the thresholds that will give the best possible f1 score.

```python
val_preds = model.predict(X_val)
best_thresholds = get_best_thresholds(y_val.values, val_preds)
for i, thresh in enumerate(best_thresholds):
    val_preds[:, i] = (val_preds[:, i] > thresh) * 1
f1_score(y_val, val_preds, average='micro')
```

```python
preds_test = model.predict(X_test)

for i, thresh in enumerate(best_thresholds):
    preds_test[:, i] = (preds_test[:, i] > thresh) * 1
```

# Transformer models

Transformer models require the data to be passed to them in two columns, column with text and a column with lists of all the target labels.

```python
# prep the data for simple transformer models
x_train = train_test[:ntrain]['clean_abstract']
y_train = train[target_cols]
X_test = train_test[ntrain:]['clean_abstract']

y_train = y_train.values.tolist()
y_train = pd.Series(y_train)

train_data = pd.concat((x_train,y_train),axis=1)
train_data.columns = ["text", "labels"]
train_data.head()
```

To implement the transformer models I will use the simpletransformers library that also allows me to use pretrained models for transfer learning.

```python
from simpletransformers.classification import (
    MultiLabelClassificationModel, MultiLabelClassificationArgs
)
import pandas as pd
import logging

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
model_type = ['bert','roberta','electra','distilbert','xlmroberta']
```

```python
model_name = ['bert-base-uncased','bert-large-uncased','roberta-base','roberta-large',
              'google/electra-base-discriminator','distilbert-base-uncased','xlm-roberta-base']
```

```python
# These models have to be ran on a gpu and require a lot of disk space so they were ran 
# separately on kaggle and google colab machines
```

```python
n=5
kf = KFold(n_splits=n, shuffle=True)

oofs = np.zeros((len(train_data),25))

for fold_, (train_index, val_index) in enumerate(kf.split(x_train,y_train)):
    print(f'\n------------- Fold {fold_ + 1} -------------')
    
  # splitting Dataframe 
    train_df = train_data.iloc[train_index]
    val_df = train_data.iloc[val_index]
    
    # Defining Model
    model = MultiLabelClassificationModel(model_type[0], model_name=model_name[0], num_labels=25,use_cuda=True, 
                                          args={'reprocess_input_data': True, 'overwrite_output_dir': True, 'num_train_epochs': 5, 'output_dir': '../tmp/', 'overwrite_output_dir':True,
                                                "cache_dir": '../tmp/cache/', "best_model_dir": "../tmp/best_model/"}
    )
  # train the model
    model.train_model(train_df)
    
    preds_val, raw_outputs = model.predict(val_df.reset_index()['text'])
    oofs[val_index] = preds_val
```

```python
predictions, raw_outputs = model.predict(X_test)
```

The predictions of the model as well as the out of fold predictions are saved in separate csv files to be used later for an ensemble model.

```python
preds_transformer = pd.DataFrame(predictions)
preds_transformer.to_csv('preds_transformer.csv', index=False)
submit_file(preds_transformer, filename='hacklive_transformer.csv')
```
